```python
from keras.datasets import mnist
import pandas as pd
import numpy as np 
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
y_train = pd.get_dummies(pd.Series(y_train)).values
y_test = pd.get_dummies(pd.Series(y_test)).values
```

```python
from keras_mlp import Mnist
```

```python
model = Mnist(x_train, y_train, x_test, y_test)
```

```python
model.build_graph()
```

```python
model.model.summary()
```

```python
model.compile_graph()
```

```python
model.train_model()
```

```python
np.argmax(model.evaluate_model(x_train[2].reshape(1,784)))
```
